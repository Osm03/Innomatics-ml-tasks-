```python
import pandas as pd

df = pd.read_csv("diminos_data.csv")
```

```python
df[['Order_date', 'Order_time']] = df['order_placed_at'].str.split(' ', expand=True)
df
```

```python
df[['Del_date', 'Del_time']] = df['order_delivered_at'].str.split(' ', expand=True)

df
```

```python
df = df.drop(['order_placed_at', 'order_delivered_at'], axis=1)

df
```
